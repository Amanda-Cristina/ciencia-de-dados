## Criando um bar plot no matplotlib
- https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py

```python
# make data
x = orientacao.index
y = orientacao.values

# plot
fig, ax = plt.subplots(figsize = (15,8))

ax.bar(x, y, width=1, edgecolor="white", linewidth=1.0)

ax.set(xlim=(0, 22), xticks=np.arange(0, 22),
       ylim=(0,7), yticks=np.arange(0, 7))

plt.show()
```

#### Para ajustar o eixo x, vamos usar o tick_params
- https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.tick_params.html

```python
# make data
x = orientacao.index
y = orientacao.values

# plot
fig, ax = plt.subplots(figsize = (15,8))

ax.bar(x, y, label="orientação")

# Rotacionando as legendas
ax.tick_params(x,labelrotation=75)

# Adicionando a legenda
ax.legend()

plt.show()
```

## Inserindo o número de orientações nos gráficos
- Para isso, vamos usar o annotate
    - https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.annotate.html
    - Com ele você consegue colocar qualquer texto no seu gráfico
    - Basta usar: `plt.annotate("TEXTO",(x,y))`
- Precisamos passar de argumentos:
    - texto que vai ser escrito
    - Posição (x,y) onde o texto vai estar

```python
# plot
fig, ax = plt.subplots(figsize = (15,8))

ax.bar(x, y, label="orientações")
ax.tick_params('x',labelrotation=75)
ax.legend()

# Colocando somente o primeiro valor no gráfico
for i in range(0,22):
    plt.annotate(y[i],
                 (i,y[i]),    # colocando a posição do elemento 
                 ha="center") # alinhamento horizontal

plt.show()
```

## Criando um scatter plot (gáfico de dispersão) usando apenas a documentação
- https://matplotlib.org/
    - Plot Types
    - scatter(x,y)

```python
# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## Rótulos no scatter plot
- https://matplotlib.org/stable/plot_types/basic/scatter_plot.html#sphx-glr-plot-types-basic-scatter-plot-py

```python
# Importar a base em excel
import pandas as pd
base = pd.read_excel('BD Analisando o engajamento no Instagram.xlsx')
```

```python
# Criando a coluna do tipo de publicação, como feito anteriormente
def tipo_para_numero(tipo):
    if(tipo == 'Foto'):
        return 1
    elif(tipo == 'Vídeo'):
        return 2
    elif(tipo == 'Reels'):
        return 3
    elif(tipo == 'IGTV'):
        return 4
    
base['NrTipo'] = base.Tipo.apply(tipo_para_numero)
```

```python
# Visualizando a base
base.head()
```

```python
# Criando o scatter plot desses dados 
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10,5))

x = base.Curtidas
y = base.Comentários

ax.scatter(x,y)
plt.show()
```

```python
# Salvando o scatter plot em uma variável
fig, ax = plt.subplots(figsize = (10,5))

x = base.Curtidas
y = base.Comentários

scatter_plot = ax.scatter(x,y,c=base.NrTipo)

plt.show()
```

```python
# Verificando essa variável
scatter_plot# Verificando essa variável
scatter_plot
```

**PathCollection**
- https://matplotlib.org/stable/api/collections_api.html?highlight=pathcollection#matplotlib.collections.PathCollection
- Uma coleção de **Paths**, como criada por, por exemplo, um scatter
    - Paths: Um módulo para lidar com as polilinhas usadas no Matplotlib. 
        - Quase todos os desenhos vetoriais usam Paths em algum lugar no pipeline de desenho.
- legend_elements()
    - Crie identificadores e rótulos de legenda para um PathCollection.
    - Retornos: handles (elementos 2D) e labels (textos dos rótulos)
    - Exemplos:
        - https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html#automatedlegendcreation

```python
# Fazendo o legend_elements()
handles, labels = scatter_plot.legend_elements()
```

**ax.legend()**
- https://matplotlib.org/stable/api/legend_api.html?highlight=legend#module-matplotlib.legend

```python
# Utilizando o legend com essa informação acima
fig, ax = plt.subplots(figsize = (10,5))

x = base.Curtidas
y = base.Comentários

scatter_plot = ax.scatter(x,y,c=base.NrTipo)

ax.legend(handles,labels)

plt.show()
```

```python
# Alterando a legenda para os nossos valores
legendas = ['Foto','Vídeo','Reels','IGTV']
fig, ax = plt.subplots(figsize = (10,5))

x = base.Curtidas
y = base.Comentários

scatter_plot = ax.scatter(x,y,c=base.NrTipo)

ax.legend(handles=handles,labels=legendas)

plt.show()
```
